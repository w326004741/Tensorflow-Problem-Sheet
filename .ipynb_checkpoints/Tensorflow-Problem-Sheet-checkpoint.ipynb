## 1. Use Tensorflow to create model

Use Tensorflow to create a model to predict the species of Iris from a flower’s sepal width, sepal length, petal width, and petal length.

```python
import tensorflow as tf
import numpy as np
from numpy import genfromtxt
import csv
import keras as kr
```

```python
# https://stackoverflow.com/questions/33633370/how-to-print-the-value-of-a-tensor-object-in-tensorflow/33633839
# https://stackoverflow.com/questions/3518778/how-to-read-csv-into-record-array-in-numpy
# return_inverse: https://docs.scipy.org/doc/numpy-1.13.0/reference/generated/numpy.unique.html

#Id = np.genfromtxt('irisdata.csv', delimiter=',',dtype=str)[1:] # Read csv file
Id = list(csv.reader(open('irisdata.csv')))[1:]# Read csv file need import csv.
#Id1 = list(csv.reader(open('irisdata.csv')))[51:53]
#Id2 = list(csv.reader(open('irisdata.csv')))[101:103]
# The inputs are four float: sl,sw,pl,pw.[:,:4]=[order:,:need]just show 4columns.
inputs = np.array(Id)[:,:4].astype(np.str)
# Outputs are initially individual strings: setosa, versicolor or virginica.output specise name.
outputs = np.array(Id)[:,4]
# return_inverse is find the unique elements of an array.
# If True, also return the indices of the unique array
# outputs_vals is output unique species name(string)
# outputs_ints is output array coding index, starting from 0,like 0,1,2....
outputs_vals, outputs_ints = np.unique(outputs, return_inverse=True)
# Encode the category integers as binary categorical variables.
# e.g. setosa(1,0,0), versicolor(0,1,0), virginica(0,0,1)
outputs_cats = kr.utils.to_categorical(outputs_ints)
#print("setosa:",outputs_cats[:5])
#print("versicolor:",outputs_cats[51:56])
#print("virginica",outputs_cats[101:106])
print(np.array(Id[:5]))
print(np.array(Id[51:55]))
print(np.array(Id[101:104]))
print('\n')
print("First 5 rows:")
print(np.array(inputs[:5]))
print(np.array(outputs[:12]))
print(np.array(outputs[51:59]))
print(np.array(outputs[101:111]))
print(np.array(outputs_vals))
print('\n')
print('Note: 0 is setosa, 1 is versicolor, 2 is virginica:')
print(np.array(outputs_ints))
print("setosa: ")
print(outputs_cats[:5])
print("versicolor: ")
print(outputs_cats[51:56])
print("virginica: ")
print(outputs_cats[101:106])
```

## 2. Split the data into training and testing

Split the data set into a training set and a testing set. You should investigate the best way to do this, and list any online references used in your notebook. If you wish to, you can write some code to randomly separate the data on the fly.

```python
# permutation: https://docs.scipy.org/doc/numpy-1.13.0/reference/generated/numpy.random.permutation.html
# split: https://docs.scipy.org/doc/numpy-1.13.0/reference/generated/numpy.array_split.html

# Randomly permute a sequence, or return a permuted range.
inds = np.random.permutation(len(inputs))
# Split an array into 2 sub-arrays: train_inds, test_inds
# That's randomly separate the data.
train_inds, test_inds = np.array_split(inds, 2)
# inputs_train = train_inds(str), outputs_train = train_inds(binary)
inputs_train, outputs_train = inputs[train_inds], outputs_cats[train_inds]
# inputs_test = test_inds(str), outputs_test = test_inds(binary)
inputs_test, outputs_test = inputs[test_inds], outputs_cats[test_inds]
# inputs_test, outputs_test
```

```python
# training set
print("Training data: ")
inputs_train[:5],'------------------------------------------------',outputs_train[:9]
```

```python
# testing set
print("Testing data: "),inputs_test[:5],'-------------------------------------------------',outputs_test[:9]
```

## 3. Train the model

Use the testing set to train your model.

```python
# Sequential: https://keras.io/getting-started/sequential-model-guide/
# Adam: https://keras.io/optimizers/#parameters-common-to-all-keras-optimizers
# relu: https://stackoverflow.com/questions/44330249/keras-using-activation-function-with-a-parameter
# Batch_size: https://stats.stackexchange.com/questions/153531/what-is-batch-size-in-neural-network
# verbose: https://keras.io/models/model/
from keras.layers import Dense, Activation
from keras.models import Sequential
# Create a neural network.
model = Sequential([
    # 16 is output_dim=hidden layer, 4=four data property(sl,sw,pl,pw).
    # 也就是从data set里读取(学习)所有的数据，然后压到16个单位.
    Dense(16, input_dim=4),
    # use ReLU function to activation funtion.
    Activation('relu'),
    # The second layer of neural network, hidden layer=3.
    Dense(3),
    # use softmax function to activation funcion.
    Activation('softmax'),
])
# define loss function use categorical_crossentropy,passed by metrics to calculate accuracy.
model.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])
# epochs=200 means have 100times train, batch_size: means how many training data each time(每次训练数据多少),
# verbose: 0, 1, or 2. Verbosity mode. 0 = silent, 1 = progress bar, 2 = one line per epoch.
model.fit(inputs_train, outputs_train, epochs=200, batch_size=5, verbose=1)
```

## 4. Test the model

Use the testing set to test your model, clearly calculating and displaying the error rate.

```python
# evaluate: https://keras.io/models/model/
# model.evaluate():Returns the loss value & metrics values for the model in test mode.
# 在测试模式下返回模型的损失值和度量值
loss, accuracy = model.evaluate(inputs_test, outputs_test, verbose=1)
print("\nFinal test set Loss:     %6.6f" % (loss))
print("Final test set Accuracy: %6.6f" % (accuracy))
```
